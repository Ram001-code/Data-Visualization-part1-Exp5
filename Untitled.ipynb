# Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Load the Dataset

```python
file_path = '.\dataset\WHR2023.csv'
df = pd.read_csv(file_path)
```

# Plot the distribution of Healthy life expectancy scores

```python
plt.hist(df['Healthy life expectancy'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Happiness Scores')
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.show()
```

# Visualize Healthy life expectancy by Region

```python
plt.figure(figsize=(12, 6))
df.groupby('Country name')['Healthy life expectancy'].mean().sort_values().plot(kind='bar')
plt.title('Average Happiness Score by Region')
plt.xlabel('Region')
plt.ylabel('Average Happiness Score')
plt.show()
```
